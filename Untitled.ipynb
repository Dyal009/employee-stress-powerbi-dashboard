```python
df["Stress_Category"] = pd.cut(
    df["Stress_Level"],
    bins=[0,40,70,100],
    labels=["Low","Medium","High"]
)
```

```python
df["Remote_Work"] = df["Remote_Work"].map({"Yes":"Remote", "No":"Office"})
```

```python
df.to_csv("employee_stress_powerbi.csv", index=False)
```
